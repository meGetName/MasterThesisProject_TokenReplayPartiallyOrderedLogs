```python
import numpy as np
import matplotlib.pyplot as plt

heuristic = [0.93, 0.76, 0.68, 23.26, 22.41, 5.29, 6.40, 0.69, 0.46, 6.75, 7.52, 9.74, 7.13 ]
worst_case = [10.23, 9.05, 7.49, 256.69, 255.24, 94.24, 64.84, 9.05, 4.65, 75.76, 284.34, 128.86, 103.04]

m, b = np.polyfit(heuristic, worst_case, 1)
lin_approx = np.poly1d([m, b])

plt.plot(heuristic, worst_case, 'ro', heuristic, lin_approx(heuristic), '--b')
plt.xlabel("Runtime heuristic")
plt.ylabel("Runtime worst case")
plt.show()
```
